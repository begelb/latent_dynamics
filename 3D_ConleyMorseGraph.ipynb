```python
import CMGDB
import math
import matplotlib
```

### Three-age class Leslie model with padding

```python
def f(x):
    theta_1 = 19.6
    theta_2 = 23.68
    theta_3 = 23.68
    return [theta_1 * x[0] + theta_2 * x[1] + theta_3 * x[2] * math.exp(-0.1 * (x[0] + x[1] + x[2])), 0.7 * x[0], 0.7 * x[1]]

''' F adapted from https://github.com/marciogameiro/CMGDB/blob/master/examples/Example_Leslie_model.ipynb '''
def F(rect):
    x0_min = rect[0]
    x1_min = rect[1]
    x0_max = rect[2]
    x1_max = rect[3]
    x2_min = rect[4]
    x2_max = rect[5]
    h_x0 = x0_max - x0_min
    h_x1 = x1_max - x1_min
    h_x2 = x2_max - x2_min
    y1 = f([x0_min, x1_min, x2_min])
    y2 = f([x0_max, x1_min, x2_min])
    y3 = f([x0_min, x1_max, x2_min])
    y4 = f([x0_max, x1_max, x2_min])

    y5 = f([x0_min, x1_min, x2_max])
    y6 = f([x0_max, x1_min, x2_max])
    y7 = f([x0_min, x1_max, x2_max])
    y8 = f([x0_max, x1_max, x2_max])

    y0_min = min([y1[0], y2[0], y3[0], y4[0], y5[0], y6[0], y7[0], y8[0]]) - h_x0
    y0_max = max([y1[0], y2[0], y3[0], y4[0], y5[0], y6[0], y7[0], y8[0]]) + h_x0
    y1_min = min([y1[1], y2[1], y3[1], y4[1], y5[1], y6[1], y7[1], y8[1]]) - h_x1
    y1_max = max([y1[1], y2[1], y3[1], y4[1], y5[1], y6[1], y7[1], y8[1]]) + h_x1
    y2_min = min([y1[2], y2[2], y3[2], y4[2], y5[2], y6[2], y7[2], y8[2]]) - h_x2
    y2_max = max([y1[2], y2[2], y3[2], y4[2], y5[2], y6[2], y7[2], y8[2]]) + h_x2
    return [y0_min, y1_min, y0_max, y1_max, y2_min, y2_max]
```

```python
lower_bounds = [0, 0, 0]
upper_bounds = [320.056, 224.040, 224.040]

subdiv_min = 12
subdiv_max = 14

morse_fname = 'morse.csv'
```

```python
morse_graph = CMGDB.MorseGraphMap(subdiv_min, subdiv_max, lower_bounds,
                                  upper_bounds, morse_fname, F)
```

```python
CMGDB.PlotMorseGraph(morse_graph, cmap=matplotlib.cm.cool)
```

```python
CMGDB.PlotMorseSets(morse_fname, cmap=matplotlib.cm.cool)
```
